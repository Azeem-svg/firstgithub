```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
## load dataset
phool = sns.load_dataset("iris")



##draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("phoolon ka plot")
# set_style
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show()
```
