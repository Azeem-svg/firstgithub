```python
### changing colors
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

## load dataset
kashti = sns.load_dataset("titanic")
kashti

##draw a line plot
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"], color = "red")
plt.show()
```

```python
### removing bars////adding palette
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

## load dataset
kashti = sns.load_dataset("titanic")
kashti

##draw a line plot
sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"], color="green", ci=None,
           palette="pastel")
plt.show()
```
